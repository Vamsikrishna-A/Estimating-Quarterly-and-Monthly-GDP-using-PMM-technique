```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_excel('Full_data.xlsx')
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

df.iloc[:, 0:28] = StandardScaler().fit_transform(df.iloc[:, 0:28])
```

```python
df.head()
```

```python
df1_X=df.drop(['GDP'],axis=1)
```

```python
df1_Y=df['GDP']
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
model = RidgeCV(alphas=np.arange(0, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')
model.fit(df1_X, df1_Y)
```

```python
model.alpha_
```

```python
model.coef_
```

```python
df1_X.columns
```

```python
coef_values=pd.DataFrame({'VarName':list(df1_X.columns),'coeff_values':list(model.coef_)},columns=['VarName','coeff_values'])
```

```python
coef_values
```

```python
coef_values['Abs(coeff_values)']=np.abs(coef_values['coeff_values'])
```

```python
coef_values
```
